```python
# Import Python Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import data into "youtube_dataset.csv"
dataset_1 = "youtube_dataset.csv"
youtubedata = pd.read_csv(dataset_1)
```

```python
# introducing  a function to sort data on the basis of number of subscribers the channel have and to form a dataset of 1000 subscribers and plot the distribution.
# sorted data will save as a csv file
   
    
def youtube(df, channeltype):
    
    df1 = df.sort_values(by=['subscribers'],ascending = False)
    df2 = df1.iloc[:1000,:]
    df2.to_csv('youtubedataset.csv')
    print(df2.groupby(channeltype)[channeltype].count().reset_index(name='Number of Channels').sort_values(['Number of Channels'], ascending=False))
    df2[channeltype].value_counts().plot(kind='bar')
    plt.ylabel('Number of Channels')
    plt.xlabel('Type of Channel ')
    plt.title('Channel Distribution') 
```

```python
# Call the function
youtube(youtubedata,'channeltype')
```
